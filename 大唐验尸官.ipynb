```python
# encoding=utf-8
import requests
import time
import pandas as pd
from bs4 import BeautifulSoup as bf
from multiprocessing import Process
url = 'http://www.ik555.net/ik258/42/42016/'
headers = {
    'User-Agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/81.0.4044.122 Safari/537.36 Edg/81.0.416.64'
}


page = requests.get(url + 'index.html', headers = headers)
page.encoding='gbk'
soup = bf(page.text,'lxml')
content = soup.select('body > div.wrap > div.d_contarin > div.ocon > dl > dd > a')

data = []
for i in content:
    data.append([i.get('title'),i.get('href')])

data[458:]
```

```python
number_of_process = 1
pool = []

def robot(threadname=0, data = None):
    if data == None:
        pass
    else:
        for i in range(len(data)):
            page = requests.get(url + data[i][1], headers = headers)
            page.encoding='gbk'
            soup = bf(page.text,'lxml')
            content = soup.select('#htmlContent')
            res = "".join(content[0].get_text().split())

            res = res.replace('datangyanshiguan。《大唐验尸官》最新章节就到艾克小说网【www.ik555.net】手机版【m.ik555.net】', '')
            with open(f'大唐验尸官-{(i+1)*50}.txt','a', encoding='utf-8') as f:
                f.writelines(['\n\n',data[i][0],'\n',res])
            print(f'{i}')
            time.sleep(0.2)

try:
    for i in range(number_of_process):
        page = 458
        pool.append(Process(target=robot, args=(i, data[page:],)))

    for t in range(number_of_process):
        pool[t].start()
        pool[t].join()
except Exception as e:
    print(e)

```
